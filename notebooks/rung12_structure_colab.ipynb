**T4 GPU**, **Run all**. ~install 12–15 min + fold ~1.5–2 h, resumable. β stays a proxy → top handles = prioritised wet-lab hypotheses.

```python
#@title Cell 5 — FOLD non-clean handles (ColabFold, MSA via MMseqs2, resumable, fail-soft)
#@markdown Only `tcr_dependent`/`anchor` handles are folded (clean → β=0 regardless). Each FASTA is
#@markdown `groove:peptide` (multimer). Rank-1 model → `<id>_{mut,wt}.pdb` for analysis.
import os, json, glob, shutil, subprocess
WORK = os.environ['RUNG12_WORK']; CFBIN = os.environ.get('COLABFOLD_BIN','colabfold_batch')
CFENV = {**os.environ, 'MPLBACKEND':'Agg'}   # FIX: Colab's inline mpl backend crashes colabfold's import
man = json.load(open('runs/rung12_pmhc/rung12_manifest.json'))
need = [h for h in man['handles'] if h.get('tier') and h['tier'] != 'clean' and 'groove' in h]
print(f'[CELL 5] {len(need)} non-clean handles x2 (mut+WT); clean skipped (β=0)')
indir = f'{WORK}/cf_in'; outdir = f'{WORK}/cf_out'; os.makedirs(indir, exist_ok=True); os.makedirs(outdir, exist_ok=True)
todo = []
for h in need:
    for tag in ('mut','wt'):
        if os.path.exists(f'{WORK}/{h["id"]}_{tag}.pdb'): continue
        src = f'{WORK}/{h["id"]}_{tag}.fasta'
        if os.path.exists(src): shutil.copy(src, f'{indir}/{h["id"]}_{tag}.fasta'); todo.append((h['id'],tag))
if todo and not os.path.exists(CFBIN):
    print('  colabfold_batch missing — SKIPPING fold (Cell 6 will use the M+P+Z fallback). Fix Cell 4 to get structure.')
elif todo:
    cmd = [CFBIN, indir, outdir, '--num-models','1','--num-recycle','3',
           '--msa-mode','mmseqs2_uniref_env','--model-type','alphafold2_multimer_v3']
    print('  running:', ' '.join(cmd))
    subprocess.run(cmd, env=CFENV)   # MPLBACKEND=Agg so colabfold's matplotlib import doesn't crash on Colab
else:
    print('  all PDBs already present (resumed).')
mapped = 0
for h in need:
    for tag in ('mut','wt'):
        dst = f'{WORK}/{h["id"]}_{tag}.pdb'
        if os.path.exists(dst): mapped += 1; continue
        hits = (sorted(glob.glob(f'{outdir}/{h["id"]}_{tag}_*rank_001*.pdb'))
                or sorted(glob.glob(f'{outdir}/{h["id"]}_{tag}*.pdb')))
        if hits: shutil.copy(hits[0], dst); mapped += 1
print(f'[CELL 5] ✓ {mapped} PDBs ready in {WORK} (resumable; re-run to continue if disconnected)')
```
